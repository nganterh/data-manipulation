```python
from tqdm import tqdm
import pandas as pd
```

```python
drug1 = pd.read_excel('drug1.xlsx', engine='openpyxl')#, index_col=0)
labsum = pd.read_csv('labsum.csv')#, index_col=0)
outcome = pd.read_csv('outcome.csv')#, index_col=0)
```

```python
tables = {'drug1': drug1, 'labsum': labsum, 'outcome': outcome}

for key, df in tqdm(tables.items()):
    df['id'] = [f'{df.at[row, "CASEID"]}{df.at[row, "VISIT"]}' for row in range(df.shape[0])]
    tables[key] = df.set_index('id')
```

```python
tables['df_merged'] = pd.DataFrame()

for i, row in tqdm(enumerate(tables['drug1'].index)):
    dict_row = tables['drug1'].loc[row].to_dict()
    #print(row, dict_row, len(dict_row))
    
    for key in ['labsum', 'outcome']:
        if row in tables[key].index:
            dict_add = tables[key].loc[row].to_dict()
            dict_row.update(dict_add)
            print('\n\n', dict_add)
    
    tables['df_merged'] = tables['df_merged'].append(dict_row, ignore_index=True)
    
    if i > 10:
        break
    
    #tables['drug1']['new_id'] = [f'{df.at[row, "CASEID"]}{df.at[row, "VISIT"]}{df.at[row, "DRUGD1"]}' for row in range(df.shape[0])]
    #tables[key] = df.set_index('id')
        
cols = [key for key in dict_row.keys()]
display(tables['df_merged'][cols])
```
