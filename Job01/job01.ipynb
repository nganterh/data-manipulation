```python
import copy
import pandas as pd
from tqdm import tqdm
```

```python
excel_book = pd.read_excel('job01.xlsx', engine='openpyxl', index_col=0, sheet_name=None)
```

```python
sheet01 = excel_book['Sheet1'].copy(deep=True)
display(sheet01)
```

```python
common_cols = tuple(sheet01.keys()[:10])
sheet03 = pd.DataFrame({col: [] for col in sheet01.keys()[:10]})

for row in tqdm(sheet01.index):
    first_values = sheet01.loc[row][:10]
    last_values = sheet01.loc[row][10:].dropna()
    
    for i, col in enumerate(common_cols):
        sheet03.at[row, col] = first_values[i]
    
    counter = 0
    
    for j, elem in enumerate(last_values):
        num = j+1
        
        if num % 2 != 0:
            sheet03.at[row, f'Change{num - counter}'] = elem
            
        else:
            sheet03.at[row, f'Time{j - counter}'] = elem
            counter += 1
    
display(sheet03)
print(common_cols)
```

```python
sheet03.to_excel('job01_done.xlsx')
```
